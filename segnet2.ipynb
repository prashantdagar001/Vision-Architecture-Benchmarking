```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input,
    Conv2D,
    BatchNormalization,
    Activation,
    MaxPooling2D,
    UpSampling2D,
)
import matplotlib.pyplot as plt
import os
import numpy as np

# Load the dataset
dataset, info = tfds.load("oxford_iiit_pet", with_info=True)
train_ds, test_ds = dataset["train"], dataset["test"]


# Define preprocessing function


def preprocess(data):

    image = tf.cast(data["image"], tf.float32)

    mask = tf.cast(data["segmentation_mask"], tf.int32)

    # Resize images and masks

    image = tf.image.resize(image, [256, 256])

    mask = tf.image.resize(mask, [256, 256], method="nearest")

    # Normalize images

    image = image / 255.0

    # Convert mask to one-hot encoding and remove extra dimension

    mask = tf.squeeze(
        tf.one_hot(mask, depth=3)
    )  # Add squeeze to remove extra dimension

    return image, mask


# Create data pipeline

BATCH_SIZE = 32

BUFFER_SIZE = 1000


train_dataset = train_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)

test_dataset = test_ds.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)


train_dataset = (
    train_dataset.cache()
    .shuffle(BUFFER_SIZE)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)

test_dataset = test_dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

# Create SegNet model
def create_segnet(input_shape, num_classes):
    def encoder_block(input_tensor, n_filters, kernel_size=3):
        x = Conv2D(n_filters, kernel_size, padding="same")(input_tensor)
        x = BatchNormalization()(x)
        x = Activation("relu")(x)
        x = Conv2D(n_filters, kernel_size, padding="same")(x)
        x = BatchNormalization()(x)
        x = Activation("relu")(x)
        return x

    def decoder_block(input_tensor, n_filters, kernel_size=3):
        x = UpSampling2D((2, 2))(input_tensor)
        x = Conv2D(n_filters, kernel_size, padding="same")(x)
        x = BatchNormalization()(x)
        x = Activation("relu")(x)
        x = Conv2D(n_filters, kernel_size, padding="same")(x)
        x = BatchNormalization()(x)
        x = Activation("relu")(x)
        return x

    # Input
    inputs = Input(shape=input_shape)

    # Encoder
    e1 = encoder_block(inputs, 64)
    p1 = MaxPooling2D((2, 2))(e1)

    e2 = encoder_block(p1, 128)
    p2 = MaxPooling2D((2, 2))(e2)

    e3 = encoder_block(p2, 256)
    p3 = MaxPooling2D((2, 2))(e3)

    e4 = encoder_block(p3, 512)
    p4 = MaxPooling2D((2, 2))(e4)

    # Bridge
    b = encoder_block(p4, 1024)

    # Decoder
    d4 = decoder_block(b, 512)
    d3 = decoder_block(d4, 256)
    d2 = decoder_block(d3, 128)
    d1 = decoder_block(d2, 64)

    # Output
    outputs = Conv2D(num_classes, 1, activation="softmax")(d1)

    model = Model(inputs=inputs, outputs=outputs)
    return model


# Create and compile model

input_shape = (256, 256, 3)

num_classes = 3


model = create_segnet(input_shape, num_classes)


# Compile model

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])


# Define callbacks

callbacks = [
    tf.keras.callbacks.ModelCheckpoint("segnet_model.keras", save_best_only=True),
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=3, min_lr=1e-6),
]


# Train the model

history = model.fit(
    train_dataset, validation_data=test_dataset, epochs=50, callbacks=callbacks
)
# Plot training history
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Plot accuracy
    ax1.plot(history.history["accuracy"])
    ax1.plot(history.history["val_accuracy"])
    ax1.set_title("Model Accuracy")
    ax1.set_ylabel("Accuracy")
    ax1.set_xlabel("Epoch")
    ax1.legend(["Train", "Validation"])

    # Plot loss
    ax2.plot(history.history["loss"])
    ax2.plot(history.history["val_loss"])
    ax2.set_title("Model Loss")
    ax2.set_ylabel("Loss")
    ax2.set_xlabel("Epoch")
    ax2.legend(["Train", "Validation"])

    plt.show()


plot_training_history(history)

# Function to visualize predictions
def visualize_prediction(image, mask, predicted_mask):
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title("Input Image")
    plt.imshow(image)

    plt.subplot(1, 3, 2)
    plt.title("True Mask")
    plt.imshow(tf.argmax(mask, axis=-1))

    plt.subplot(1, 3, 3)
    plt.title("Predicted Mask")
    plt.imshow(tf.argmax(predicted_mask, axis=-1))

    plt.show()


# Visualize some predictions
for image, mask in test_dataset.take(3):
    predicted_mask = model.predict(image)

    # Visualize the first image in the batch
    visualize_prediction(image[0].numpy(), mask[0].numpy(), predicted_mask[0])

# Evaluate model on test dataset
test_loss, test_accuracy = model.evaluate(test_dataset)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Calculate IoU (Intersection over Union) metric
def calculate_iou(y_true, y_pred):
    y_true = tf.argmax(y_true, axis=-1)
    y_pred = tf.argmax(y_pred, axis=-1)

    class_ious = []

    for class_id in range(num_classes):
        true_class = tf.cast(y_true == class_id, tf.float32)
        pred_class = tf.cast(y_pred == class_id, tf.float32)

        intersection = tf.reduce_sum(true_class * pred_class)
        union = tf.reduce_sum(true_class) + tf.reduce_sum(pred_class) - intersection

        iou = (intersection + 1e-7) / (union + 1e-7)
        class_ious.append(iou)

    return tf.reduce_mean(class_ious)


# Calculate IoU on test dataset
total_iou = 0
num_batches = 0

for image_batch, mask_batch in test_dataset:
    pred_batch = model.predict(image_batch)
    batch_iou = calculate_iou(mask_batch, pred_batch)
    total_iou += batch_iou
    num_batches += 1

mean_iou = total_iou / num_batches
print(f"\nMean IoU: {mean_iou:.4f}")

# Save model predictions on test set
def save_predictions(dataset, num_samples=5):
    output_dir = "predictions"
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for idx, (image, mask) in enumerate(dataset.take(num_samples)):
        predicted_mask = model.predict(image)

        for i in range(image.shape[0]):  # Loop through batch
            # Save original image
            plt.imsave(f"{output_dir}/sample_{idx}_{i}_image.png", image[i].numpy())

            # Save true mask
            plt.imsave(
                f"{output_dir}/sample_{idx}_{i}_true_mask.png",
                tf.argmax(mask[i], axis=-1),
                cmap="viridis",
            )

            # Save predicted mask
            plt.imsave(
                f"{output_dir}/sample_{idx}_{i}_pred_mask.png",
                tf.argmax(predicted_mask[i], axis=-1),
                cmap="viridis",
            )


save_predictions(test_dataset)

# Function to perform inference on a single image
def predict_single_image(image_path):
    # Load and preprocess the image
    image = tf.io.read_file(image_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.cast(image, tf.float32)
    image = tf.image.resize(image, [256, 256])
    image = image / 255.0
    image = tf.expand_dims(image, 0)  # Add batch dimension

    # Predict
    prediction = model.predict(image)
    prediction = tf.argmax(prediction[0], axis=-1)

    # Visualize
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Input Image")
    plt.imshow(tf.squeeze(image))

    plt.subplot(1, 2, 2)
    plt.title("Predicted Mask")
    plt.imshow(prediction, cmap="viridis")

    plt.show()


# Example usage of single image prediction
# predict_single_image('path_to_your_image.jpg')

# Save model architecture and weights
model.save("segnet_full_model.h5")
model.save_weights("segnet_weights.weights.h5")

# Save model architecture as JSON
model_json = model.to_json()
with open("segnet_architecture.json", "w") as json_file:
    json_file.write(model_json)

print("Model and weights saved successfully!")

# Print model summary and training parameters
print("\nModel Training Summary:")
print(f"Number of training samples: {len(train_dataset) * BATCH_SIZE}")
print(f"Number of test samples: {len(test_dataset) * BATCH_SIZE}")
print(f"Number of classes: {num_classes}")
print(f"Input shape: {input_shape}")
print(f"Batch size: {BATCH_SIZE}")
print(f"Initial learning rate: {model.optimizer.learning_rate.numpy()}")

# Additional metrics for each class
def calculate_class_metrics(dataset):
    class_correct = np.zeros(num_classes)
    class_total = np.zeros(num_classes)

    for images, masks in dataset:
        predictions = model.predict(images)
        pred_classes = tf.argmax(predictions, axis=-1)
        true_classes = tf.argmax(masks, axis=-1)

        for class_id in range(num_classes):
            mask = true_classes == class_id
            correct = pred_classes[mask] == class_id
            class_correct[class_id] += tf.reduce_sum(tf.cast(correct, tf.float32))
            class_total[class_id] += tf.reduce_sum(tf.cast(mask, tf.float32))

    class_accuracy = class_correct / (class_total + 1e-7)
    return class_accuracy


print("\nPer-class Accuracy:")
class_accuracies = calculate_class_metrics(test_dataset)
for i, acc in enumerate(class_accuracies):
    print(f"Class {i}: {acc:.4f}")
```
